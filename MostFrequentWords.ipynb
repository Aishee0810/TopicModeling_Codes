```
data = pd.read_excel("/content/drive/My Drive/Topic Modeling/Dataset_Combined.xlsx")
```

```
combo = list(data['Combined'])
```

```
combo_com = []
punctuations = '''!()-[]{};:'"\,<>./?@$%^&*~।|॥…=৷—'''
for comment in combo: 
    comment1 = comment.replace("Unavailable","")
    for char in comment:
      if char in punctuations:
        comment1=comment1.replace(char," ")
    combo_com.append(comment1.replace("\n",""))
```

```
words=[]
for comment in combo_com:
  words.append(comment.split(' '))
```

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_eng = set(nltk.corpus.stopwords.words('english'))
```

```
f=open("/content/drive/My Drive/Topic Modeling/BengaliStopwords.txt",encoding="utf8")
stopwords_beng=f.read()
```

```
words_without_empty=[]
for comment in words:
  for element in comment:
    element = element.lower()
    if(element!='' and element not in stopwords_beng and element not in stopwords_eng):
       words_without_empty.append(element)
```

```
from banglakit import lemmatizer as lem
from banglakit.lemmatizer import BengaliLemmatizer
from bangla_stemmer.stemmer import stemmer
```

```
def lemmatize_stemming_bengali(text):
  stmr = stemmer.BanglaStemmer().stem(BengaliLemmatizer().lemmatize(text,pos=lem.POS_VERB))
  return SnowballStemmer("english").stem(WordNetLemmatizer().lemmatize(stmr, pos='v'))
```

```
lemmatized=[]
for comment in words_without_empty:
  lemmatized.append(lemmatize_stemming_bengali(comment))
```

```
def wordListToFreqDict(wordlist):
    wordfreq = [wordlist.count(p) for p in wordlist]
    return dict(list(zip(wordlist,wordfreq))) 
```

```
td_freq = (wordListToFreqDict(words_without_empty))
#td_freq
```
