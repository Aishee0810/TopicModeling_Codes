```
import pandas as pd
```

```
data = pd.read_excel("/content/drive/My Drive/Topic Modeling/Voting_TopicNo.xlsx")
data
```

```
post_number = list(data['Post _Number'])
true_label = list(data['final'])
tfidf5 = list(data['TFIDF5_TopicNo'])
```

```
true_posts = [[] for i in range(5)]

for i in range(len(post_number)):
  if true_label[i] == 'A':
    true_posts[0].append(post_number[i])
  elif true_label[i] == 'B':
    true_posts[1].append(post_number[i])
  elif true_label[i] == 'C':
    true_posts[2].append(post_number[i])
  elif true_label[i] == 'D':
    true_posts[3].append(post_number[i])
  elif true_label[i] == 'E':
    true_posts[4].append(post_number[i])
```

```
tfidf5_posts = [[] for i in range(5)]

for i in range(len(post_number)):
  if tfidf5[i] == 'Topic: 0':
    tfidf5_posts[0].append(post_number[i])
  elif tfidf5[i] == 'Topic: 1':
    tfidf5_posts[1].append(post_number[i])
  elif tfidf5[i] == 'Topic: 2':
    tfidf5_posts[2].append(post_number[i])
  elif tfidf5[i] == 'Topic: 3':
    tfidf5_posts[3].append(post_number[i])
  elif tfidf5[i] == 'Topic: 4':
    tfidf5_posts[4].append(post_number[i])
```

```
map = [[] for i in range(len(true_posts))]

for i in range(len(true_posts)):
  for j in range(len(tfidf5_posts)):
    map[i].append((str(j+1),len(set(true_posts[i]).intersection(set(tfidf5_posts[j])))))
```

```
map #each list represents the actual class(A,B,C,D,E)
```

```
for i in range(len(map)):
  print(max(map[i],key=lambda j:j[1]))
```

```
tf2combi = []
combi2 = []
for i in range(0,len(tfidf5_posts)-1):
  for j in range(i+1,len(tfidf5_posts)):
    tf2combi.append(list(set(tfidf5_posts[i]).union(set(tfidf5_posts[j]))))   #for combining two topic lists
    combi2.append(str(i+1)+str(j+1))
```

```
map2 = [[] for i in range(len(true_posts))]

for i in range(len(true_posts)):
  for j in range(len(tf2combi)):
    map2[i].append((combi2[j],len(set(true_posts[i]).intersection(set(tf2combi[j])))))
```

```
map2 # first elements in tuple represent topic combinations
```

```
for i in range(len(map2)):
  print(max(map2[i],key=lambda j:j[1]))
```

```
tf3combi = []
combi3 = []
for i in range(0,len(tfidf5_posts)-2):
  for j in range(i+1,len(tfidf5_posts)-1):
    for k in range(j+1,len(tfidf5_posts)):
      tf3combi.append(list(set(tfidf5_posts[i]).union(set(tfidf5_posts[j]),set(tfidf5_posts[k])))) #for combining three topic lists
      combi3.append(str(i+1)+str(j+1)+str(k+1))
```

```
map3 = [[] for i in range(len(true_posts))]

for i in range(len(true_posts)):
  for j in range(len(tf3combi)):
    map3[i].append((combi3[j],len(set(true_posts[i]).intersection(set(tf3combi[j])))))
```

```
map3 # first elements in tuple represent topic combinations
```

```
for i in range(len(map3)):
  print(max(map3[i],key=lambda j:j[1]))
```

```
final3list = [[] for i in range(len(map))] #merging the combinations

for i in range(0,len(map)):
  final3list[i].extend(map[i])
  final3list[i].extend(map2[i])
  final3list[i].extend(map3[i])
```

```
true_class = ['A','B','C','D','E']

for i in range(len(final3list)):
  print(true_class[i]+" -> "+str(max(final3list[i],key=lambda j:j[1])))
```

```
import random
```

```
dist = [144,120,72,48,96]

sample = []

for i in range(len(dist)):
  sample.append(random.sample(true_posts[i],dist[i])) # creating random samples for aiming to form proportionate distribution of topics
```

```
for i in range(0,len(sample)):
  print(len(sample[i]))
```

```
sample_posts = []

for i in range(len(sample)):
  for j in range(len(sample[i])):
    sample_posts.append(sample[i][j])
```

```
post_number, true_label,tfidf20
```

```
sample_tf = [[] for i in range(20)]


for i in range(len(post_number)):
  if post_number[i] in sample_posts:
    for j in range(20):
      if tfidf20[i] == 'Topic: '+str(j):
        sample_tf[j].append(post_number[i])
```

```
for i in range(0,len(sample_tf)):
  print(len(sample_tf[i]))
```

```
sample_map20 = [[] for i in range(len(sample))]

for i in range(len(sample)):
  for j in range(len(sample_tf)):
    sample_map20[i].append((str(j+1),len(set(sample[i]).intersection(set(sample_tf[j])))))
```
